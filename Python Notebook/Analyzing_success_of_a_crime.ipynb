```python
from Data_Utility import filterData,getCount,sortValue
```

```python
completed_crimes = filterData(data_to_tranform,7,'COMPLETED')
```

```python
sortFelony = sortValue(filterData(getCount(completed_crimes,list([9,11,8]),'FELONY'),8,'FELONY'),'FELONY',False)
sortVio = sortValue(filterData(getCount(completed_crimes,list([9,11,8]),'VIOLATION'),8,'VIOLATION'),'VIOLATION',False)
sortMis = sortValue(filterData(getCount(completed_crimes,list([9,11,8]),'MISDEMEANOR'),8,'MISDEMEANOR'),'MISDEMEANOR',False)
mergedData = sortFelony.merge(sortMis.merge(sortVio, left_on=[9,11], right_on=[9,11], how='outer'), left_on=[9,11], right_on=[9,11], how='outer')
```
