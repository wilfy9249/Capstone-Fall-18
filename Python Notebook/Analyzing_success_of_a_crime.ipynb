```python
crimes_original = getCleanedData()
```

```python
#Drop unnecessary features
data_to_tranform = crimes_original.drop([0,1,2,3,4,6,12,13,14],axis=1)
```

```python
crime_filter = data_to_tranform[7].value_counts(dropna=False)
```

```python
from data_utility import filterData,getCount,sortValue
completed_crimes = filterData(data_to_tranform,7,'COMPLETED')
```

```python
felony_val = 'FELONY'
mis_val = 'MISDEMEANOR'
vio_val = 'VIOLATION'

only_completed_felony = filterData(completed_crimes,8,felony_val)
data = only_completed_felony.drop([5,7,10],axis = 1)
fel_data = getCount(data,list([9,11,8]),felony_val)
fel_data.columns= ['BOROUGHS', 'PREMISES', 'TYPE', felony_val]

only_completed_mis = filterData(completed_crimes,8,mis_val)
data = only_completed_mis.drop([5,7,10],axis = 1)
mis_data = getCount(data,list([9,11,8]),mis_val)

only_completed_vio = filterData(completed_crimes,8,vio_val)
data = only_completed_vio.drop([5,7,10],axis = 1)
vio_data = getCount(data,list([9,11,8]),vio_val)
```

```python
fel_data[[felony_val]] = fel_data[[felony_val]].apply(pd.to_numeric)
```

```python
fel_pivot = fel_data.pivot_table(values = [felony_val], index = ['PREMISES'], columns = ['BOROUGHS'])
fel_pivot.insert(5,'Total',fel_pivot.sum(axis=1))
```

```python
sorted_fel = fel_pivot.reindex(fel_pivot['Total'].sort_values(ascending=False).index)
```
